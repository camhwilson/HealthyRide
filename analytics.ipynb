```python
from weekday_analytics.plotting_funcs import plot_all
from neighborhood_analytics.plotting_funcs import Neighborhood_Plot


from weekday_analytics.weekday import group_weekdaydict_by_station

from trip import Trip

import json
```

```python
with open('data.json') as f:
  json_list = json.load(f)
```

```python
print(json_list[0])
```

```python
#convert list of dicts to list of objects called "Trips"
Trip_list = []
for dict in json_list:
    li = list(dict.values())
    Trip_list.append(Trip(*li))
```

```python
#test er out
print(Trip_list[420].toname)
print(Trip_list[420].starttime)
print(Trip_list[420].weekday)
print(Trip_list[420]._id)
```

```python
clensed_Trip_list = [i for i in Trip_list if type(i.weekday) is int]
print(len(clensed_Trip_list))
```

```python
neighborhood_weekday_dict = group_weekdaydict_by_station(clensed_Trip_list)

```

```python
plot_all(neighborhood_weekday_dict)
```
